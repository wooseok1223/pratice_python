```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

face_cascade = cv2.CascadeClassifier(
    './opencv-master/opencv-master/data/haarcascades/haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier(
    './opencv-master/opencv-master/data/haarcascades/haarcascade_smile.xml')

image = cv2.imread('./FinalData2/test3.jpg')
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(grayImage, 1.03,  5)
smile = smile_cascade.detectMultiScale(grayImage, 1.01, 10)
print(faces)
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
    
    smile_image_gray = grayImage[y:y+h, x:x+w]
    smile_image_color = image[y:y+h, x:x+w]
    
    faces_in_smile = face_cascade.detectMultiScale(smile_image_gray)
    
    
    for (xf,yf,wf,hf) in faces_in_smile:
        cv2.rectangle(smile_image_color,(xf,yf),(xf+wf,yf+hf),(255,0,0),2)

cv2.rectangle(image, ((0,image.shape[0] -25)), 
              (270, image.shape[0]), (255,255,255), -1);
cv2.putText(image, "!", (0,image.shape[0] -10), 
            cv2.FONT_HERSHEY_TRIPLEX, 0.5,  (0,0,0), 1);

plt.figure(figsize=(12,12))
plt.imshow(image, cmap='gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
import cv2
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

face_cascade = cv2.CascadeClassifier(
    './opencv-master/opencv-master/data/haarcascades/haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier(
    './opencv-master/opencv-master/data/haarcascades/haarcascade_smile.xml')

img = cv2.imread('./FinalData2/test3.jpg')
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(grayImage, 1.03,5)
smile = smile_cascade.detectMultiScale(grayImage, 1.01, 10)

imgNum  = 0
for (x,y,w,h) in faces:
    cropped = img[y - int(h / 4):y + h + int(h / 4), x - int(w / 4):x + w + int(w / 4)]
    #  
    cv2.imwrite("./final_data/thumbnail" + str(imgNum) + ".jpg", cropped)
    imgNum += 1

    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0),2)
    smile_image_gray = grayImage[y:y+h, x:x+w]
    smile_image_color = grayImage[y:y+h, x:x+w]
    
    faces_in_smile = face_cascade.detectMultiScale(smile_image_gray)
    for (xf,yf,wf,hf) in faces_in_smile:
        cv2.rectangle(smile_image_color,(xf,yf),(xf+wf,yf+hf),(255,0,0),2)

cv2.imshow('Image view', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
