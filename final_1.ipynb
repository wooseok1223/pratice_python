```python
import time #코드를 실행하는데 딜레이를 걸어주는 모듈
from selenium import webdriver #가상 드라이브를 이용하여 html 정보를 받아온다.
from selenium.webdriver.common.keys import Keys #셀레니움의 키를 인식하는 모듈
from bs4 import BeautifulSoup
import urllib.request #크롤링 모듈이지만 이미지 저장을 위해 사용
import urllib

chrome_path = "C:/pratice/chromedriver_win32/chromedriver" #자신의 크롬드라이브의 경로를 지정
base_url = "https://ohou.se/cards/feed?utf8=✓&query=모던" #접속하고자하는 url
driver = webdriver.Chrome(chrome_path) #크롬드라이브를 실행
driver.implicitly_wait(5) #명시적대기 크롬드라이브가 완전히 켜질때까지 5초의 대기를 걸어줌
driver.get(base_url) #가상드라이브를 이용하여 접속하고자하는 url에 접속

elem = driver.find_element_by_tag_name('body') # 스크롤을 내리기위하여 바디 태그를 검색
pagedowns = 1 #페이지 기본값
targetpage = 100#목표 페이지
imglist = [] #사진의 경로를 담을 리스트

temp1 = []
temp2 = []

while pagedowns < targetpage: #현재 페이지값이 목표페이지 값보다 작을때 수행
    elem.send_keys(Keys.PAGE_DOWN) #스크롤 한번!!!
    time.sleep(0.1) #스크롤 한번 내리는동안 페이지의 정보를 가져오기 위하여 1초의 대기시간을 걸어준다.
    pagedowns += 1 #페이지를 증가시킨다.
    print(pagedowns)

soup = BeautifulSoup(driver.page_source, 'html.parser')    
a = soup.find_all('img')
try:
    for i in a:
        a = i.get('src')
        imglist.append(a)
except:
    pass    
    
    
search = 'snapshots'
search2 = 'projects'
for j in imglist:
    if search in str(j):
        temp1.append(j)
    elif search2 in str(j):
        temp2.append(j)

print(temp1)
print(temp2)

# # 드라이버를 종료한다.
# driver.close()

##################################################################################
```

```python
import pandas as pd
df1 = pd.DataFrame(temp1)
df1 = df1.rename(columns = {0 : 'src'})
df1 = df1.drop_duplicates()

df2 = pd.DataFrame(temp2)
df2 = df2.rename(columns = {0 : 'src'})
df2 = df2.drop_duplicates()

final = pd.concat([df1,df2])

FinalSrc = []
for i in final['src']:
    FinalSrc.append(i)
    
print(FinalSrc)
final
```

```python
import urllib.request

for j in range((len(FinalSrc))): #리스트의 길이만큼 반복한다.
    img_name = 'A'+str(j)+'.png' #사진의 이름을 A0부터 리스트의 길이까지 지정
    urllib.request.urlretrieve(FinalSrc[j], "./Furniture/modern/" + img_name) #img폴더에 사진을 저장
```
