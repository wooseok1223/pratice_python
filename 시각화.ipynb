```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(tips['total_bill'],bins=20)
ax.set_title('Histogram of Total Bill')
ax.set_xlabel("frequency")
ax.set_ylabel("Total Bill")
```

```python
# 두 양적 데이터의 단순 상관관계 조사 => 산포도부터 ㄱㄱ

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(tips['total_bill'], tips['tip'])
ax.set_title('Scatterplot of Total Bill vs Tip')
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
```

```python
# 성별, tip의 관계
# 질적데이터와 양적데이터의 비교 => boxplot 으로 비교

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot([tips[tips["sex"]=="Female"]["tip"], tips[tips["sex"]=="Male"]["tip"]], labels=["Female", "Male"])
ax.set_title('Boxplot of Tips by Sex')
ax.set_xlabel("sex")
ax.set_ylabel("tip")
```

```python
def recode_sex(sex):
    if sex == "Female":
        return 0
    else:
        return 1
    
tips["sex_color"] = tips['sex'].apply(recode_sex)
tips.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x=tips['total_bill'], y=tips['tip'], c=tips['sex_color'], s=tips['size']*10, alpha=0.5)
ax.set_title('Scatterplot of Total Bill vs Tip')
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
```

```python
# wordcloud

#!pip install wordcloud
#!pip install feedparser #newspaper, konlpy
#!pip install newspaper3k

from wordcloud import WordCloud
import feedparser
import newspaper
```

```python
rss_link = 'http://www.khan.co.kr/rss/rssdata/economy.xml'
feeds = feedparser.parse(rss_link)
feeds

links = [entry['link'] for entry in feeds['entries']]
links


news_text = ""
for link in links:
    article = newspaper.Article(link, language="ko")
    article.download()
    article.parse()
    news_text += article.text
    
news_text[:100]
len(news_text)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
text = open('data/wordcloud/constitution.txt').read()
text
wordcloud = WordCloud(max_font_size=30).generate(text)
wordcloud.words_

plt.figure(figsize=(12, 12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
```

```python
text = open('data/wordcloud/alice.txt').read()
text
stopwords = set(STOPWORDS)
stopwords.add('said')

wc = WordCloud(mask=alice_mask, stopwords=stopwords, background_color='white')
wc = wc.generate(text)
wc.words_

plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#!pip install konlpy
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
```

```python
ko_text = kolaw.open('constitution.txt').read()
ko_text
```

```python
t = Twitter()
token_ko = t.nouns(ko_text)
token_ko

stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안']
token_ko = [each_word for each_word in token_ko if each_word not in stop_words]
token_ko

ko = nltk.Text(token_ko, name="대한민국 헌법")
data = ko.vocab().most_common(500)
tmp_data = dict(data)

wc = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color="white").generate_from_frequencies(tmp_data)
plt.figure(figsize=(16, 8))
plt.imshow(wc)
```
